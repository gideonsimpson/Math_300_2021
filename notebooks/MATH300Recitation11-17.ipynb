# Problem 4.5.1(c)

Use Romberg integration to compute $R_{3, 3}$ for the following integral: $\newline \newline$
$$
\int_{0}^{0.35}\frac{2}{x^{2} - 4}dx.
\newline
$$

## Solution:

To compute the exact value of this integral, we use partial fraction decomposition: $\newline \newline$
$$
\int_{0}^{0.35}\frac{2}{x^{2} - 4}dx = \frac{1}{2}\int_{0}^{0.35}\frac{dx}{x - 2} - \frac{1}{2}\int_{0}^{0.35}\frac{dx}{x + 2}.
\newline
$$
Then, the evaluation is straightforward: $\newline \newline$
$$
\begin{align*}
    \frac{1}{2}\int_{0}^{0.35}\frac{dx}{x - 2} - \frac{1}{2}\int_{0}^{0.35}\frac{dx}{x + 2} &= \frac{1}{2}\ln|x - 2|\Big|_{0}^{0.35} - \frac{1}{2}\log|x + 2|\Big|_{0}^{0.35} \\
    &= \frac{1}{2}\Big(\ln(1.65) - \ln(2)\Big) - \frac{1}{2}\Big(\ln(2.35) - \ln(2)\Big) \\
    &= \frac{1}{2}\Big(\ln(1.65) - \ln(2.35)\Big) \\
    &= \frac{1}{2}\Big(\ln\frac{1.65}{2.35}\Big) \\
    &\approx -0.17682002.
\end{align*}
$$

To compute this integral with Romberg integration, we need to use the composite trapezoidal rule and Richardson extrapolation (this is the definition of Romberg integration, of course). First, we recall the approximation made by the composite trapezoidal rule: $\newline \newline$
$$
\int_{a}^{b}f(x)dx = \frac{h}{2}\Big[f(a) + 2\sum_{j = 1}^{n - 1}f(x_{j}) + f(b)\Big] + \mathcal{O}(h^{2}),
\newline
$$
where $a = x_{0} < x_{1} < \dots < x_{n - 1} < x_{n} = b$ are equally spaced nodes.

Also recall, that for a numerical scheme with truncation errors with only even powers of $h$, i.e., for numerical schemes that have the form $\newline \newline$
$$
M = N_{1}(h) + K_{1}h^{2} + K_{2}h^{4} + K_{3}h^{6} + \dots,
\newline
$$
Richardson extrapolation makes the approximation $\newline \newline$
$$
N_{j}(h) = N_{j - 1}\Big(\frac{h}{2}\Big) + \frac{N_{j - 1}(h/2) - N_{j - 1}(h)}{4^{j - 1} - 1}
\newline
$$
for all $j \geq 2$, where $N_{1}(h)$ is the usual approximation (without extrapolation) using step size $h$.

In the notation of the section, we compute $\newline \newline$
$$
R_{1, 1} = \frac{b - a}{2}[f(a) + f(b)], \\
R_{k, 1} = \frac{1}{2}\Big[R_{k - 1, 1} + h_{k - 1}\sum_{i = 1}^{2^{k - 2}}f(a + (2i - 1)h_{k})\Big],
\newline
$$
and then use extrapolation to compute $\newline \newline$
$$
R_{k, j} = R_{k, j - 1} + \frac{1}{4^{j - 1} - 1}(R_{k, j - 1} - R_{k - 1, j - 1}), \;\; k \geq j.
\newline
$$
We will fill out the table

| $k$ | $\mathcal{O}(h_{k}^{2})$ | $\mathcal{O}(h_{k}^{4})$ | $\mathcal{O}(h_{k}^{6})$ |
| --- | --- | --- | --- |
| $1$ | $R_{1, 1}$ |  |  |
| $2$ | $R_{2, 1}$ | $R_{2, 2}$ |  |
| $3$| $R_{3, 1}$ | $R_{3, 2}$ | $R_{3, 3}$ |

To start, we compute $R_{1, 1}$: $\newline \newline$
$$
R_{1, 1} = \frac{h_{1}}{2}\Big[f(a) + f(b)\Big] = \frac{0.35 - 0}{2}\Bigg[\frac{2}{0^{2} - 4} + \frac{2}{0.35^{2} - 4}\Bigg] \approx -0.17776435.
\newline
$$
Notice that this is relatively close to the true value already (of course, we have done the trapezoidal method!), but this serves as a good way to check that we are not making arithmetic mistakes along the way. Now, we compute $R_{2, 1}$: $\newline \newline$
$$
R_{2, 1} = \frac{1}{2}\Big[R_{1, 1} + h_{1}f(a + h_{2})\Big] \approx \frac{1}{2}\Big[-0.17776435 + 0.35\frac{2}{0.175^2 - 4}\Big] \approx -0.17705726.
\newline
$$
From these, we can use extrapolation to obtain $R_{2, 2}$: $\newline \newline$
$$
R_{2, 2} = R_{2, 1} + \frac{1}{4^{2 - 1} - 1}\Big(R_{2, 1} - R_{1, 1}\Big) \approx -0.17705726 + \frac{1}{3}(-0.17705726 - (-0.17776435)) \approx -0.17682157.
\newline
$$
Now, we compute the entries of the third row. $\newline \newline$
$$
R_{3, 1} = \frac{1}{2}\Big[R_{2, 1} + h_{2}\sum_{i = 1}^{2^{3 - 2}}f(a + (2i - 1)h_{3})\Big] \approx \frac{1}{2}\Bigg[-0.17705726 + 0.175\Bigg(\frac{2}{0.0875^2 - 4} + \frac{2}{0.2625^{2} - 4}\Bigg)\Bigg] \approx −0.17687941, \\
R_{3, 2} = R_{3, 1} + \frac{1}{4^{2 - 1} - 1}(R_{3, 1} - R_{2, 1}) \approx −0.17687941 + \frac{1}{3}(−0.17687941 - (-0.17705726)) \approx −0.17682012.
\newline
$$
Finally, we have all the information we need to compute $R_{3, 3}$: $\newline \newline$
$$
R_{3, 3} = R_{3, 2} + \frac{1}{15}(R_{3, 2} - R_{2, 2}) \approx −0.17682012 + \frac{1}{15}(−0.17682012 - (-0.17682157)) \approx −0.17682002.
\newline
$$
We see that $R_{3, 3}$ is accurate to (at least) 8 decimal places, which seems reasonable since $\Big(\frac{h}{4}\Big)^{6} \approx 4.4 \times 10^{-7}$.

# Problem 4.5.14

In Exercise 24 of Section 1.1, a Maclaurin series was integrated to approximate $\text{erf}(1)$, where $\text{erf}(1)$ is the normal distribution error function defined by $\newline \newline$
$$
\text{erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x}e^{-t^{2}}dt.
\newline
$$
(The error function can be interpreted as the probability of $Y \in [0, x]$ if $Y \sim \mathcal{N}(0, \frac{1}{\sqrt{2}})$). Approximate $\text{erf}(1)$ to within $10^{-7}$.

## Solution: 

We need to approximate the integral $\newline \newline$
$$
\frac{2}{\sqrt{\pi}}\int_{0}^{1}e^{-t^{2}}dt.
\newline
$$
We begin constructing the Romberg table. We have $\newline \newline$
$$
R_{1, 1} = \frac{1}{2}\Big[f(0) + f(1)\Big] \approx 0.77174333
\newline
$$
for the first row. For the second row, we have $\newline \newline$
$$
R_{2, 1} = \frac{1}{2}\Big[R_{1, 1} + f(0.5)\Big] \approx 0.82526296, \\
R_{2, 2} = R_{2, 1} + \frac{1}{3}\Big[R_{2, 1} - R_{1, 1}\Big] \approx 0.84310283.
\newline
$$
Continuing in this manner, we produce the table

| $1$ | $0.77174333$ |  |  |  |  |  |
| $2$ | $0.82526296$ | $0.84310283$ |  |  |  |  |
| $3$| $0.83836778$ | $0.84273605$ | $0.84271160$ |  |  |  |

Notice that the entries in the bottom right of the table all agree out to $8$ digits, so we are clearly converging to the solution with error less than or equal to $10^{-7}$. Indeed, if we held more digits in our calculation, we would find that $\newline \newline$


# Problem 4.7.1(c)

Approximate the following integral using Gaussian quadrature with $n = 2$ and compare your results to the exact value of the integral: $\newline \newline$
$$
\int_{0}^{0.35}\frac{2}{x^{2} - 4}dx.
\newline
$$

## Solution: 

Recall that the exact solution of this integral to 8 decimal places is $-0.17682002$.

Gaussian quadrature calls for us to make a change of variables so that our interval of integration is $[-1, 1]$ instead of $[0, 0.35]$. We accomplish this by making the change of variable $\newline \newline$
$$
x = \frac{1}{2}\Big[0.35t + 0.35\Big],
\newline
$$
or equivalently, $\newline \newline$
$$
t = \frac{2x - 0.35}{0.35}.
\newline
$$
Then, the integral becomes $\newline \newline$
$$
\int_{-1}^{1}\frac{2}{\Big(\frac{0.35t + 0.35}{2}\Big)^{2} - 4}\cdot 0.175dt
\newline
$$
From the analysis done in the book, we know that for $n = 2$, the coefficients and roots to use for Gaussian quadrature are $1$ (for both coefficients) and $\pm \frac{\sqrt{3}}{3}$ for the roots. That is, an approximation to our integral is $\newline \newline$
$$
\begin{align*}
\int_{-1}^{1}\frac{2}{\Big(\frac{0.35t + 0.35}{2}\Big)^{2} - 4}\cdot 0.175dt &\approx \frac{2}{\Big(\frac{0.35\frac{\sqrt{3}}{3} + 0.35}{2}\Big)^{2} - 4}\cdot 0.175 + \frac{2}{\Big(\frac{0.35\frac{-\sqrt{3}}{3} + 0.35}{2}\Big)^{2} - 4}\cdot 0.175 \\
&\approx −0.17681899
\end{align*}
\newline
$$
This is (almost shockingly) accurate for how little work was done; in fact, we have accuracy on the order of $10^{-4}$ with a clever substitution and only 2 (only 2!) function evaluations. This is the power of Gaussian quadrature.
